```python
import requests
from termcolor import cprint

data = {'domain': open('domain.pddl').read(),
        'problem': open('problem.pddl').read()}

resp = requests.post('http://solver.planning.domains/solve', 
                     verify=False, json=data).json()

plan = ('\n'.join([act['name'] for act in resp['result']['plan']]))
cost = (resp['result']['length'])
output = (resp['result']['output'])

#------------------------------------------------------------------

print("COST/NUMBER OF STEPS:", cost)
print("----------------------------")

#------------------------------------------------------------------

#Delete '()' symbols from plan and get a list of steps divided in actions-items 
plan = plan.replace('(', '').replace(')', '')
steps = plan.split("\n")
actions = [i.split() for i in steps]

print(plan)
print("")

#Classify actions or insert them into functions
print("ACTION TYPES:")
for line in actions:
    #print(line)
    if line[0] == 'move':
        cprint(" MOVE ","green",attrs=["reverse","bold"], end = ' ')
    if line[0] == 'pick':
        cprint(" PICK ","yellow",attrs=["reverse","bold"], end = ' ')
    if line[0] == 'drop':
        cprint(" DROP ","blue",attrs=["reverse","bold"], end = ' ')
```
